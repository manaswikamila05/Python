## Cleaning the dataset - I

The company rating (C_rating) has `NaNs` and is of `object` type. Clean the column by imputing the nulls and converting the column to numeric type

### Check if nulls are the only impurities present here and why the column is of 'object' type

```python
# Let's directly convert it to a numeric type and see what results we get
# For NaNs we'll coerce the errors later, for now let's check if there are some other types of errors or not
pd.to_numeric(df.C_Rating)
```

Okay, based on the above error, apart from NaNs, it seems that some of the data contain impurities in them.  **This is because this might not be the only erroneous pattern** Let's inspect if indeed this is the case

**ValueError** observed in the stack trace. Pattern is of the form x.x

```python
# One logic to check which types of errors exist can be to inspect the length of each of the entries in the C_Rating column
# Ideally the number should be the of the type " *.*" where * indicates a numeric digit. Hence the length should be 3
# If the length is more than 3, then it might contain an error.

for i in (df[~df['C_Rating'].isna()]['C_Rating']):
    if len(i)>3:
        print(i)
```

The rating column contains a few characters along with the numbers that deonte the rating. These characters need to be removed.

**Bonus method to find the rows which show error**

```python
errors = []

for i in (df['C_Rating']):
   try: 
    pd.to_numeric(i)
   except ValueError:
    errors.append(i)
    
print(errors)
```

### Convert the column to the correct datatype

```python
# With the nulls imputed, we need to remove the impurities 
# Let's write a custom function to do the same
```

```python
print(errors)
```

```python
# Extract only numbers and decimal point
```

```python
def cleaner(str):
    clean_str="".join(filter(lambda x: (x in "0123456789."), str))
    return float(clean_str)
```

```python
cleaner('4.4*')
```

```python
cleaner('2*+.9')
```

**Explanation** - The function `cleaner` takes a string and filters out all the numeric digits and the dot symbol, joins them and then converts the given string to a float value

```python
df['C_Rating'] = df['C_Rating'].apply(cleaner)
```

```python
df2['Salary Estimate'] = df2['J_Salary'].apply(lambda x : re.sub('[^0-9-]','',x))
df2.head()
```
